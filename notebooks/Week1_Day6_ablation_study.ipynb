```python

# DAY 6 - ABLATION STUDY
# Member 2 - Context Specialist

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

```

```python
# LOAD DATASET (Contextual Dataset from Day 5)


df = pd.read_csv("contextual_feature_engineered_ai4i.csv")

print("Dataset Loaded")
print(df.shape)

```

```python
# TARGET VARIABLE


y = df["Machine failure"]

```

```python

# MODEL 1: SENSOR FEATURES ONLY


sensor_features = [
    "Air temperature [K]",
    "Process temperature [K]",
    "Rotational speed [rpm]",
    "Torque [Nm]",
    "Tool wear [min]",
    "Type",
    "temp_difference"
]

X_sensor = df[sensor_features]

X_train, X_test, y_train, y_test = train_test_split(
    X_sensor,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

model_sensor = RandomForestClassifier(random_state=42)

model_sensor.fit(X_train, y_train)

pred_sensor = model_sensor.predict(X_test)

sensor_report = classification_report(y_test, pred_sensor, output_dict=True)


```

```python
# MODEL 2: SENSOR + CONTEXT FEATURES


context_features = [
    "ambient_temperature",
    "factory_load",
    "humidity",
    "temp_gap",
    "load_torque_interaction",
    "humidity_temp_interaction"
]

X_context = df[sensor_features + context_features]

X_train, X_test, y_train, y_test = train_test_split(
    X_context,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

model_context = RandomForestClassifier(random_state=42)

model_context.fit(X_train, y_train)

pred_context = model_context.predict(X_test)

context_report = classification_report(y_test, pred_context, output_dict=True)
```

```python

# RESULT COMPARISON TABLE


results = pd.DataFrame({
    "Model": ["Sensor Only", "Sensor + Context"],
    "Precision (1)": [
        sensor_report["1"]["precision"],
        context_report["1"]["precision"]
    ],
    "Recall (1)": [
        sensor_report["1"]["recall"],
        context_report["1"]["recall"]
    ],
    "F1 Score (1)": [
        sensor_report["1"]["f1-score"],
        context_report["1"]["f1-score"]
    ]
})

print("\nAblation Study Results:\n")
print(results)
```
