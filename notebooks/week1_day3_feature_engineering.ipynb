```python
import pandas as pd

# Load dataset

df = pd.read_csv("../content/ai4i2020.csv")

# Preview dataset

df.head()
```

```python
# Create Temperature Difference Feature

df["temp_difference"] = (
    df["Process temperature [K]"]
    - df["Air temperature [K]"]
)

df[["temp_difference"]].head()
```

```python
# Summary Statistics

df[["temp_difference"]].describe()
```

```python
# Encode Product Type

df["Type"] = df["Type"].map({
    "L": 0,
    "M": 1,
    "H": 2
})

df["Type"].value_counts().to_frame()
```

```python
# Remove Identifier Columns

df = df.drop(
    ["UDI", "Product ID"],
    axis=1
)

df.head()
```

```python
# Machine Failure Distribution

df["Machine failure"].value_counts().to_frame()
```

```python
# Machine Failure Percentage

(
    df["Machine failure"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
    .to_frame(name="Percentage")
)
```

```python
# Correlation with Target

corr = df.corr(numeric_only=True)

corr["Machine failure"]\
    .sort_values(ascending=False)\
    .to_frame(name="Correlation")
```

```python
# Final Dataset Shape

pd.DataFrame({
    "Rows": [df.shape[0]],
    "Columns": [df.shape[1]]
})
```

```python
# Preview Processed Dataset

df.head()
```

```python
# Save Processed Dataset

df.to_csv(
    "../content/processed_ai4i.csv",
    index=False
)

print("Processed dataset saved successfully.")
```
