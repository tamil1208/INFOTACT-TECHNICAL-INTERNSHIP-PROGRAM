```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("contextual_feature_engineered_ai4i.csv")

print(df.shape)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x="Machine failure",
    y="temp_gap",
    data=df
)
plt.title("Temperature Gap vs Machine Failure")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(
    df["factory_load"],
    bins=20,
    kde=True
)
plt.title("Factory Load Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x="Machine failure",
    y="humidity",
    data=df
)
plt.title("Humidity vs Machine Failure")
plt.show()
```

```python
context_cols = [
    "ambient_temperature",
    "factory_load",
    "humidity",
    "temp_gap",
    "load_torque_interaction",
    "humidity_temp_interaction",
    "Machine failure"
]

plt.figure(figsize=(8,6))
sns.heatmap(
    df[context_cols].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Context Feature Correlation")
plt.show()
```


OBSERVATIONS

1. Machines with larger temperature gaps show slightly higher failure tendencies.

2. Factory load varies considerably across operating conditions, indicating the importance of contextual workload information.

3. Humidity levels alone show weak relationships with failures but may influence machine behavior when combined with temperature.

4. Contextual features demonstrate measurable interactions with operational variables and can contribute additional insights beyond sensor-only analysis.

5. These visualizations support the inclusion of environmental and operational context in predictive maintenance workflows.


CONCLUSION

This analysis explored the role of contextual features in understanding machine operating conditions and failure behavior. Additional environmental and operational variables such as ambient temperature, factory load, humidity, temperature gap, and interaction-based features were successfully engineered and analyzed.

The visualizations indicate that contextual variables provide supplementary information about machine behavior beyond traditional sensor measurements. In particular, temperature gap and load-related interaction features demonstrated observable relationships with operational conditions and machine failures.

These findings suggest that contextual information can support a more comprehensive understanding of machine health and may contribute valuable insights for future predictive maintenance and risk assessment workflows.
