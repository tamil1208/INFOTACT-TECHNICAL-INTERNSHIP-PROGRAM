```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/ai4i2020.csv")

df.head()
```

```python
print(df.shape)

df.info()
```

```python
df["temp_difference"] = (
    df["Process temperature [K]"]
    - df["Air temperature [K]"]
)
```

```python
df["Type"] = df["Type"].map({
    "L": 0,
    "M": 1,
    "H": 2
})
```

```python
df = df.drop(
    ["UDI", "Product ID"],
    axis=1
)
```

```python
df["power_proxy"] = (
    df["Rotational speed [rpm]"]
    * df["Torque [Nm]"]
)
```

```python
df["wear_torque_ratio"] = (
    df["Tool wear [min]"]
    /
    (df["Torque [Nm]"] + 1)
)
```

```python
df["temp_stress_index"] = (
    df["temp_difference"]
    * df["Torque [Nm]"]
)
```

```python
df["rpm_torque_interaction"] = (
    df["Rotational speed [rpm]"]
    * df["Torque [Nm]"]
)
```

```python
df["wear_category"] = pd.cut(
    df["Tool wear [min]"],
    bins=[0,50,150,300],
    labels=["Low","Medium","High"]
)
```

```python
df["wear_category"] = (
    df["wear_category"]
    .cat.codes
)
```

```python
new_features = [
    "power_proxy",
    "wear_torque_ratio",
    "temp_stress_index",
    "rpm_torque_interaction",
    "wear_category"
]

df[new_features].head()
```

```python
corr = df.corr(numeric_only=True)

corr["Machine failure"]\
.sort_values(ascending=False)
```

```python
df[new_features].describe()
```

```python
df.to_csv(
    "/content/feature_engineered_ai4i.csv",
    index=False
)

print("Feature engineered dataset saved.")
```
